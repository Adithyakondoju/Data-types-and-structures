2. Explain the difference between mutable and immutable data types with examples
- Mutable Data Types
A mutable object can be changed "in-place." When you modify it, you are updating the data stored at the same memory address rather than creating a new one.
- eg: Lists ([1, 2]), Dictionaries ({'a': 1}), Sets.
-Immutable data types
-An immutable object cannot be changed once it is created. If you try to "modify" it, the computer actually creates a brand new object in a different memory location with the new value.
- eg: Strings ("hello"), Tuples ((1, 2)), Integers, Booleans.

3.  What are the main differences between lists and tuples in Python?
- the primary distinction between a list and a tuple lies in their mutability and their intended semantic purpose.
- A list is a mutable, dynamic array—defined with square brackets []—that allows you to modify, add, or remove elements in-place, making it the go-to structure for collections of data that will change during a program's execution.
- In contrast, a tuple is an immutable sequence—defined with parentheses ()—that cannot be altered once created. From a technical standpoint, tuples are more memory-efficient and slightly faster because their fixed size allows Python to allocate memory more precisely.

4. Describe how dictionaries store data
- The Key is "Hashed": When you give the dictionary a key (like the word "Apple"), Python runs it through a hash function. This turns the word into a specific number, like 502.
- The Address is Found: That number tells Python exactly where in memory to store the data. It’s like saying, "Everything labeled 'Apple' always goes into Locker #502."
- Instant Retrieval: When you want to find the value for "Apple", Python doesn't look at "Banana" or "Cherry." it just runs the math, gets the number 502, and jumps straight to that spot.

5. Why might you use a set instead of a list in Python
- We would choose a set over a list primarily when you need to ensure uniqueness or perform high-speed membership testing. While lists are ordered collections that allow duplicates, sets are unordered collections of unique elements based on mathematical set theory.

6. What is a string in Python, and how is it different from a list
- a string is a sequence of characters used to represent text (like "Hello"). While it may look similar to a list of characters, the fundamental difference lies in mutability and purpose.
- The Core Difference: Mutability
The most important distinction is that strings are immutable, whereas lists are mutable.

- Strings: Once a string is created, you cannot change any character within it. If you try to change a letter, Python will throw an error. To "change" a string, you must create a completely new one.

- Lists: You can freely change, add, or remove elements at any time.

7. How do tuples ensure data integrity in Python
- tuples ensure data integrity primarily through immutability. This means that once a tuple is created and stored in memory, its state cannot be altered by any part of the program.

8. What is a hash table, and how does it relate to dictionaries in Python
- A hash table is a data structure that organizes data into key-value pairs to allow for nearly instantaneous data retrieval. In Python, the dictionary is the high-level implementation of a hash table.

9.  Can lists contain different data types in Python
- Yes, Python lists are heterogeneous, meaning they can store a variety of different data types simultaneously within a single list.

- Unlike many other programming languages (like C++ or Java) where an array is usually "typed" to hold only one kind of data, a Python list is essentially a collection of pointers to objects. Each pointer can point to a different type of object in memory.

10. Explain why strings are immutable in Python
- In Python, strings are immutable primarily for reasons involving memory efficiency, security, and system stability.

11. What advantages do dictionaries offer over lists for certain tasks
- A dictionary is better than a list when you need:Faster Lookups: Dictionaries find items instantly ($O(1)$) using a unique key, whereas lists must check every item one-by-one ($O(n)$).
- Data Mapping: Dictionaries allow you to label data with meaningful names (like "email": "test@abc.com") instead of just a numeric index (like 0, 1, 2).Unique Keys: Dictionaries automatically prevent duplicate keys, which is ideal for managing unique records like User IDs or IP addresses.
- Memory Efficiency for Sparsity: Dictionaries are better for storing data associated with non-sequential IDs (e.g., storing data for ID #1 and ID #10,000 without needing 9,998 empty slots in between).

13. How do sets handle duplicate values in Python
- In Python, sets handle duplicate values by automatically discarding them during both creation and modification. This behavior is built into the set's core architecture, which is based on a hash table.

14.  How does the “in” keyword work differently for lists and dictionaries
- The 'in' keyword behaves differently for lists and dictionaries primarily due to their underlying data structures and how they store and access data.

16. What is a nested dictionary, and give an example of its use case
- A nested dictionary is a dictionary that contains other dictionaries as its values. In programming, particularly in Python, this allows you to represent complex, hierarchical data structures where information is organized into levels.

- Real-World Use Case: E-Commerce Product Catalog
One of the most common use cases for nested dictionaries is representing JSON data from APIs or complex product catalogs.
Imagine an e-commerce site that needs to store products. Each product has fixed attributes (name, price), but also category-specific specifications that are better organized in sub-groups.

17.  Describe the time complexity of accessing elements in a dictionary
- In Python, the time complexity for accessing an element in a dictionary (by its key) is 1$O(1)$ on average.2 This is known as constant time complexity.Whether your dictionary has 10 items or 10 million items, the time it takes to retrieve a value remains virtually the same because dictionaries use a data structure called a hash table.

19.  Why are dictionaries considered unordered, and how does that affect data retrieval
- The concept of "order" in Python dictionaries has an interesting history. While modern Python dictionaries appear to maintain order, the underlying data structure is technically designed as an unordered mapping.

20. Explain the difference between a list and a dictionary in terms of data retrieval.
- 1. Access Method: Indices vs. Keys
The most visible difference is the "handle" you use to grab your data.

- List (Index-based): You retrieve elements based on their position. A list is an ordered sequence, so you must know the numerical index (starting at 0) to get a specific value.

- Dictionary (Key-based): You retrieve elements based on a label. A dictionary is a mapping, meaning you use a unique key (like a string or ID) to get a value.
